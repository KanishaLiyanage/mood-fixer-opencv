```python
import os
import cv2
import keras
import cvzone
import random
import numpy as np
import tkinter as tk
import customtkinter as ctk
from playsound import playsound
from keras.utils import np_utils
from keras.models import load_model
from keras.preprocessing import image
```

```python
emotions_classifier = load_model('emotions_classifier.h5')
```

```python
face_classifier = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
emotion_labels = ['Angry', 'Disgust', 'Fear', 'Happy', 'Neutral', 'Sad', 'Surprise']
```

```python
emoji_size = 150
robot_size = 120
message_size = (250, 200)
emoji_position = [470, 310]
robot_position = [8, 200]
message_position = [10, 10]

robot_img = cv2.imread('images/robotic.png')
```

```python
ctk.set_appearance_mode("System")  #Modes: system (default), light, dark
ctk.set_default_color_theme("blue")  #Themes: blue (default), dark-blue, green

root = ctk.CTk()  #create CTk window like you do with the Tk window
root.geometry("800x480")

emoji_size = 150
robot_size = 120
message_size = (250, 200)
emoji_position = [470, 310]
robot_position = [8, 200]
message_position = [10, 10]

robot_img = cv2.imread('images/robotic.png')

def beginMoodFixer():
    print("Bot Button Pressed!")
    cap = cv2.VideoCapture(0)
    #cap.set(3, 1280)
    #cap.set(4, 720)
    while True:
        _, frame = cap.read()
        labels = []
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        faces = face_classifier.detectMultiScale(gray, scaleFactor=1.2, minNeighbors=5)
    
        black_bar = cv2.rectangle(frame, (7, 450), (250, 465), (0,0,0), -1)
        exit_text = 'Press "ESC" to exit from this window'
        exit_text_position = (10, 460)
        cv2.putText(frame, exit_text, exit_text_position, cv2.FONT_HERSHEY_SIMPLEX, 0.4, (0, 0, 255), 1)
    
        robot_img = cv2.imread('images/robotic.png', cv2.IMREAD_UNCHANGED)
        robot_img = cv2.resize(robot_img, (robot_size, robot_size))
        frame = cvzone.overlayPNG(frame, robot_img, robot_position)

        for (x, y, w, h) in faces:
            cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 255), 2)
            roi_gray = gray[y:y+h, x:x+w]
            roi_gray = cv2.resize(roi_gray, (48, 48), interpolation=cv2.INTER_AREA)
        
            if np.sum([roi_gray])!=0:
                roi = roi_gray.astype('float')/255.0
                roi = keras.utils.img_to_array(roi)
                roi = np.expand_dims(roi, axis=0)

                prediction = emotions_classifier.predict(roi)[0]
                label = emotion_labels[prediction.argmax()]
                label_position = (x, y-10)
                #label = label + str(prediction)
                mood = label + " Mood"
                cv2.putText(frame, mood, label_position, cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
            
                if label == 'Angry':
                    message = cv2.imread("messages/Angry.png", cv2.IMREAD_UNCHANGED)
                    message = cv2.resize(message, message_size)
                    frame = cvzone.overlayPNG(frame, message, message_position)
                
                    emoji = cv2.imread("emojis/Angry.png", cv2.IMREAD_UNCHANGED)
                    emoji = cv2.resize(emoji, (emoji_size, emoji_size))
                    frame = cvzone.overlayPNG(frame, emoji, emoji_position)
                
                elif label == 'Disgust':
                    message = cv2.imread("messages/Disgust.png", cv2.IMREAD_UNCHANGED)
                    message = cv2.resize(message, message_size)
                    frame = cvzone.overlayPNG(frame, message, message_position)
                
                    emoji = cv2.imread("emojis/Disgust.png", cv2.IMREAD_UNCHANGED)
                    emoji = cv2.resize(emoji, (emoji_size, emoji_size))
                    frame = cvzone.overlayPNG(frame, emoji, emoji_position)
                
                elif label == 'Fear':
                    message = cv2.imread("messages/Fear.png", cv2.IMREAD_UNCHANGED)
                    message = cv2.resize(message, message_size)
                    frame = cvzone.overlayPNG(frame, message, message_position)
                
                    emoji = cv2.imread("emojis/Fear.png", cv2.IMREAD_UNCHANGED)
                    emoji = cv2.resize(emoji, (emoji_size, emoji_size))
                    frame = cvzone.overlayPNG(frame, emoji, emoji_position)
                
                elif label == 'Happy':
                    message = cv2.imread("messages/Happy.png", cv2.IMREAD_UNCHANGED)
                    message = cv2.resize(message, message_size)
                    frame = cvzone.overlayPNG(frame, message, message_position)
                
                    emoji = cv2.imread("emojis/Happy.png", cv2.IMREAD_UNCHANGED)
                    emoji = cv2.resize(emoji, (emoji_size, emoji_size))
                    frame = cvzone.overlayPNG(frame, emoji, emoji_position)
                
                elif label == 'Neutral':
                    message = cv2.imread("messages/Neutral.png", cv2.IMREAD_UNCHANGED)
                    message = cv2.resize(message, message_size)
                    frame = cvzone.overlayPNG(frame, message, message_position)
                
                    emoji = cv2.imread("emojis/Neutral.png", cv2.IMREAD_UNCHANGED)
                    emoji = cv2.resize(emoji, (emoji_size, emoji_size))
                    frame = cvzone.overlayPNG(frame, emoji, emoji_position)
                
                elif label == 'Sad':
                    message = cv2.imread("messages/Sad.png", cv2.IMREAD_UNCHANGED)
                    message = cv2.resize(message, message_size)
                    frame = cvzone.overlayPNG(frame, message, message_position)
                
                    emoji = cv2.imread("emojis/Sad.png", cv2.IMREAD_UNCHANGED)
                    emoji = cv2.resize(emoji, (emoji_size, emoji_size))
                    frame = cvzone.overlayPNG(frame, emoji, emoji_position)
                
                elif label == 'Surprise':
                    message = cv2.imread("messages/Surprise.png", cv2.IMREAD_UNCHANGED)
                    message = cv2.resize(message, message_size)
                    frame = cvzone.overlayPNG(frame, message, message_position)
                
                    emoji = cv2.imread("emojis/Surprise.png", cv2.IMREAD_UNCHANGED)
                    emoji = cv2.resize(emoji, (emoji_size, emoji_size))
                    frame = cvzone.overlayPNG(frame, emoji, emoji_position)
                
            else:
                cv2.putText(frame, 'No Face detected!', (30, 80), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
            
        cv2.imshow('Mood Fixer V1.0', frame)
    
        if cv2.waitKey(1) & 0xFF == 27: #27 for esc key
            break

    cap.release()
    cv2.destroyAllWindows()
   

def beginMusic():
    print("Music Button Pressed!")
    cap = cv2.VideoCapture(0)
    #cap.set(3, 1280)
    #cap.set(4, 720)
    while True:
        _, frame = cap.read()
        labels = []
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        faces = face_classifier.detectMultiScale(gray, scaleFactor=1.2, minNeighbors=5)
    
        black_bar = cv2.rectangle(frame, (7, 450), (250, 465), (0,0,0), -1)
        exit_text = 'Press "ESC" to exit from this window'
        exit_text_position = (10, 460)
        cv2.putText(frame, exit_text, exit_text_position, cv2.FONT_HERSHEY_SIMPLEX, 0.4, (0, 0, 255), 1)

        for (x, y, w, h) in faces:
            cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 255), 2)
            roi_gray = gray[y:y+h, x:x+w]
            roi_gray = cv2.resize(roi_gray, (48, 48), interpolation=cv2.INTER_AREA)
        
            if np.sum([roi_gray])!=0:
                roi = roi_gray.astype('float')/255.0
                roi = keras.utils.img_to_array(roi)
                roi = np.expand_dims(roi, axis=0)

                prediction = emotions_classifier.predict(roi)[0]
                label = emotion_labels[prediction.argmax()]
                label_position = (x, y-10)
                mood = label + " Mood"
                cv2.putText(frame, mood, label_position, cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
            
                if label == 'Angry':
                    emoji = cv2.imread("emojis/Angry.png", cv2.IMREAD_UNCHANGED)
                    emoji = cv2.resize(emoji, (emoji_size, emoji_size))
                    frame = cvzone.overlayPNG(frame, emoji, emoji_position)

                    music_dir = "C:\\Users\\Kanisha Liyanage\\Documents\\Deep Learning Projects\\mood-fixer-opencv\\music\\for_angry\\"
                    files = os.listdir(music_dir)
                    music = random.choice(files)
                    os.startfile(os.path.join(music_dir, music))

                elif label == 'Neutral':
                    emoji = cv2.imread("emojis/Neutral.png", cv2.IMREAD_UNCHANGED)
                    emoji = cv2.resize(emoji, (emoji_size, emoji_size))
                    frame = cvzone.overlayPNG(frame, emoji, emoji_position)

                    music_dir = "C:\\Users\\Kanisha Liyanage\\Documents\\Deep Learning Projects\\mood-fixer-opencv\\music\\for_neutral\\"
                    files = os.listdir(music_dir)
                    music = random.choice(files)
                    os.startfile(os.path.join(music_dir, music))
                
                elif label == 'Sad':
                    emoji = cv2.imread("emojis/Sad.png", cv2.IMREAD_UNCHANGED)
                    emoji = cv2.resize(emoji, (emoji_size, emoji_size))
                    frame = cvzone.overlayPNG(frame, emoji, emoji_position)

                    music_dir = "C:\\Users\\Kanisha Liyanage\\Documents\\Deep Learning Projects\\mood-fixer-opencv\\music\\for_sad\\"
                    files = os.listdir(music_dir)
                    music = random.choice(files)
                    os.startfile(os.path.join(music_dir, music))
                
            else:
                cv2.putText(frame, 'No Face detected!', (30, 80), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
            
        cv2.imshow('Mood Fixer V1.0', frame)
    
        if cv2.waitKey(1) & 0xFF == 27: #27 for esc key
            break

    cap.release()
    cv2.destroyAllWindows()


welcome = ctk.CTkLabel(master=root, text="Welcome to MoodFixer V1.0", font=("Poppins", 35))
welcome.place(relx=0.5, rely=0.15, anchor=tk.CENTER)

aboutText = ctk.CTkLabel(master=root, text="This is a system that identifying user's emotions in real-time.", font=("Poppins", 12))
aboutText.place(relx=0.5, rely=0.3, anchor=tk.CENTER)

botLabel = ctk.CTkLabel(master=root, text="Fix your mood with our MoodBot", font=("Poppins", 12))
botLabel.place(relx=0.4, rely=0.5, anchor=tk.CENTER)
botBtn = ctk.CTkButton(master=root, text="Begin", command=beginMoodFixer)
botBtn.place(relx=0.65, rely=0.5, anchor=tk.CENTER)

musicLabel = ctk.CTkLabel(master=root, text="Fix your mood with a music", font=("Poppins", 12))
musicLabel.place(relx=0.4185, rely=0.65, anchor=tk.CENTER)
musicBtn = ctk.CTkButton(master=root, text="Begin", command=beginMusic)
musicBtn.place(relx=0.65, rely=0.65, anchor=tk.CENTER)

# Create an Exit button.
exitBtn = ctk.CTkButton(root, text = "Exit", command = root.destroy)
exitBtn.place(relx=0.5, rely=0.85, anchor=tk.CENTER)

root.mainloop()
```
