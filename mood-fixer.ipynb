```python
import cv2
import keras
import numpy as np
from keras.utils import np_utils
from keras.models import load_model
from keras.preprocessing import image
```

```python
emotions_classifier = load_model('emotions_classifier.h5')
```

```python
face_classifier = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

emotion_labels = ['Angry', 'Disgust', 'Fear', 'Happy', 'Neutral', 'Sad', 'Surprise']

cap = cv2.VideoCapture(0)

#cap.set(3, 1280)
#cap.set(4, 720)

while True:
    _, frame = cap.read()
    labels = []
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    faces = face_classifier.detectMultiScale(gray)
    
    exit_text = 'Press "ESC" to exit from this window'
    exit_text_position = (10, 460)
    cv2.putText(frame, exit_text, exit_text_position, cv2.FONT_HERSHEY_SIMPLEX, 0.4, (0, 0, 255), 1)

    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 255), 2)
        roi_gray = gray[y:y+h, x:x+w]
        roi_gray = cv2.resize(roi_gray, (48, 48), interpolation=cv2.INTER_AREA)
        
        if np.sum([roi_gray])!=0:
            roi = roi_gray.astype('float')/255.0
            roi = keras.utils.img_to_array(roi)
            roi = np.expand_dims(roi, axis=0)

            prediction = emotions_classifier.predict(roi)[0]
            label = emotion_labels[prediction.argmax()]
            label_position = (x, y-10)
            cv2.putText(frame, label, label_position, cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
            
            if label == 'Angry':
                suggestion_text = 'Hey! why are you in angry mood? just cool down mate.'
                suggestion_text_position = (20, 20)
                cv2.putText(frame, suggestion_text, suggestion_text_position, cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
                img = cv2.imread("emojis/Angry.png")
                size = 150
                img = cv2.resize(img, (size, size))
                img2gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
                ret, mask = cv2.threshold(img2gray, 1, 255, cv2.THRESH_BINARY)
                emoji_roi = frame[-size-10:-10, -size-10:-10]
                emoji_roi[np.where(mask)] = 0
                emoji_roi += img
                #img = cv2.resize(img, (150, 150))
                #cv2.imshow('Angry', img)
                
            elif label == 'Disgust':
                suggestion_text = 'All is disgust when a man leaves his own nature and does what is unfit.'
                suggestion_text_position = (20, 20)
                cv2.putText(frame, suggestion_text, suggestion_text_position, cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
                img = cv2.imread("emojis/Disgust.png")
                size = 150
                img = cv2.resize(img, (size, size))
                img2gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
                ret, mask = cv2.threshold(img2gray, 1, 255, cv2.THRESH_BINARY)
                emoji_roi = frame[-size-10:-10, -size-10:-10]
                emoji_roi[np.where(mask)] = 0
                emoji_roi += img
                
            elif label == 'Fear':
                suggestion_text = 'It’s okay to be scared. Being scared means you’re about to do something really, really brave.'
                suggestion_text_position = (20, 20)
                cv2.putText(frame, suggestion_text, suggestion_text_position, cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
                img = cv2.imread("emojis/Fear.png")
                size = 150
                img = cv2.resize(img, (size, size))
                img2gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
                ret, mask = cv2.threshold(img2gray, 1, 255, cv2.THRESH_BINARY)
                emoji_roi = frame[-size-10:-10, -size-10:-10]
                emoji_roi[np.where(mask)] = 0
                emoji_roi += img
                
            elif label == 'Happy':
                suggestion_text = 'I am also happy, because of you are in happy mood.'
                suggestion_text_position = (20, 20)
                cv2.putText(frame, suggestion_text, suggestion_text_position, cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
                img = cv2.imread("emojis/Happy.png")
                size = 150
                img = cv2.resize(img, (size, size))
                img2gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
                ret, mask = cv2.threshold(img2gray, 1, 255, cv2.THRESH_BINARY)
                emoji_roi = frame[-size-10:-10, -size-10:-10]
                emoji_roi[np.where(mask)] = 0
                emoji_roi += img
                
            elif label == 'Neutral':
                suggestion_text = 'It doesn\'t matter to stay calm and neutral all the time.'
                suggestion_text_position = (20, 20)
                cv2.putText(frame, suggestion_text, suggestion_text_position, cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
                img = cv2.imread("emojis/Neutral.png")
                size = 150
                img = cv2.resize(img, (size, size))
                img2gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
                ret, mask = cv2.threshold(img2gray, 1, 255, cv2.THRESH_BINARY)
                emoji_roi = frame[-size-10:-10, -size-10:-10]
                emoji_roi[np.where(mask)] = 0
                emoji_roi += img
                
            elif label == 'Sad':
                suggestion_text = 'Don\'t be sad mate. just relax and forget it.'
                suggestion_text_position = (20, 20)
                cv2.putText(frame, suggestion_text, suggestion_text_position, cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
                img = cv2.imread("emojis/Sad.png")
                size = 150
                img = cv2.resize(img, (size, size))
                img2gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
                ret, mask = cv2.threshold(img2gray, 1, 255, cv2.THRESH_BINARY)
                emoji_roi = frame[-size-10:-10, -size-10:-10]
                emoji_roi[np.where(mask)] = 0
                emoji_roi += img
                
            elif label == 'Surprise':
                suggestion_text = 'Most things in life come as a surprise.'
                suggestion_text_position = (20, 20)
                cv2.putText(frame, suggestion_text, suggestion_text_position, cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
                img = cv2.imread("emojis/Surprise.png")
                size = 150
                img = cv2.resize(img, (size, size))
                img2gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
                ret, mask = cv2.threshold(img2gray, 1, 255, cv2.THRESH_BINARY)
                emoji_roi = frame[-size-10:-10, -size-10:-10]
                emoji_roi[np.where(mask)] = 0
                emoji_roi += img
                
        else:
            cv2.putText(frame, 'No Face detected!', (30, 80), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
            
    cv2.imshow('Mood Fixer V1.0', frame)
    
    if cv2.waitKey(1) & 0xFF == 27: #27 for esc key
        break

cap.release()
cv2.destroyAllWindows()
```
